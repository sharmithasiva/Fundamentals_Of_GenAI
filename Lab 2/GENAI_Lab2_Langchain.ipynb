```python
from langchain_community.llms.fake import FakeListLLM
from langchain_core.prompts import PromptTemplate

# ---------------------------------
# Fake LLM (Offline, No API Key)
# ---------------------------------
llm = FakeListLLM(
    responses=[
        "LangChain is a framework used to build applications powered by language models."
    ]
)

# ---------------------------------
# Prompt Template
# ---------------------------------
prompt = PromptTemplate(
    input_variables=["context", "question"],
    template="""
Use the context below to answer the question.

Context:
{context}

Question:
{question}

Answer:
"""
)

# ---------------------------------
# Create Runnable Chain (NEW WAY)
# ---------------------------------
chain = prompt | llm

# ---------------------------------
# Context
# ---------------------------------
context = """
LangChain is a framework for developing applications powered by large language models.
It provides tools for prompt templates, chains, memory, and integrations.
"""

# ---------------------------------
# Question
# ---------------------------------
question = "What is LangChain?"

# ---------------------------------
# Run Chain
# ---------------------------------
result = chain.invoke({
    "context": context,
    "question": question
})

print("Answer:", result)
```
